## Importação de dados

A classe `DataFrame` serve para que possamos retornar uma tabela no Python.
Abaixo temos o seguinte:

* A variável `dados` armazena um dicionário (tudo que está dentro de `{}`).
* O dicionário é composto por várias chaves (`nome`, `idade`, `altura`, `peso`, `cidade`).
* Cada chave possui uma lista de valores (os dados que correspondem a cada chave) relacionada a uma coluna do DataFrame.

Com a variável `dados` elaborada, instanciamos os valores através da variável `df`, que, por sua vez, recebe o construtor `pd.DataFrame()`.

```python
dados = {
    'nome': ['Lucas', 'Joseph', 'Carl', 'Ana', None, 'Erick', 'Marcus'],
    'idade': [23, 35, None, 23, 40, '45', '22'],
    'altura': [1.65, 1.80, 1.75, 1.65, None, 1.70, 1.85],
    'peso': [55, 80, 77, 55, 90, None, 93],
    'estado': ['SP', 'MG', 'SP', 'SC', 'PR', 'MS', 'GO']
}

df = pd.DataFrame(dados)

df
```

Também podemos retornar uma coluna em específico, ao invés de um DataFrame completo, usando o tipo `Series`.

* Instanciamos uma nova variável.
* Atribuímos o valor de `df['nome_da_coluna']` (onde `nome_da_coluna` é o nome da coluna que queremos retornar).

```python
nome_series = df['nome']

nome_series
```

## Análise de dados

Para casos em que o DataFrame tem um grande volume de dados, podemos observar somente os primeiros 5 registros utilizando o método `.head()`.

```python
df.head()
```

Ou pode-se optar por usar o método `.tail()` para visualizar os últimos registros do DataFrame.

```python
df.tail()
```

Podemos validar os dados usando uma comparação booleana.

Se adicionarmos uma condição para comparação e referenciarmos a coluna, podemos obter um resultado booleano.

```python
df['altura'] >= 1.70

```

Com essa comparação booleana, podemos aplicar filtros chamando a variável `df` e passando a condição para que nos retorne apenas o que desejamos.

```python
df[df['altura'] >= 1.70]
```

Voltando ao DataFrame completo, podemos observar que valores ausentes recebem a sigla NaN (Not a Number).

Essas ausências ou faltas de valores são uma constante de ponto flutuante estabelecida pelo padrão IEEE (Instituto de Engenheiros Eletricistas e Eletrônicos).

Para uma análise de dados utilizando Pandas, não podemos ter campos com ausência de valores.

Nesses casos, realizamos os próximos passos de análise e limpeza de dados para que não resulte em nenhum erro em nossa análise.

Começamos com a análise dos dados usando o método `.info()`.

O método `.info()` irá nos mostrar:
* Valores não nulos
* Tipos de dados
* Tamanho da memória utilizada

```python
df.info()
```

O método `.describe()` gera um resumo estatístico rápido das colunas numéricas (e opcionalmente categóricas) de um DataFrame, facilitando a análise exploratória de dados.

```python
df.describe()
```

### Ordenação de dados

Podemos ordenar nossos resultados usando o método `.sort_values()`.

```python
df.sort_values(by='altura')
```

Para ordenarmos do maior para o menor, usamos o parâmetro `ascending=False`.

```python
df.sort_values(by='altura', ascending=False)
```

## Tratamento de dados

Se observar a coluna `idade`, ela contém valores não numéricos (tipo `object`).

Se tivermos que fazer alguma operação matemática, precisamos converter esses dados para tipos numéricos (`int` ou `float`).

```python
df['idade'] = pd.to_numeric(df['idade'], errors='coerce')
df
```

### Adicionando colunas.

Podemos adicionar novas colunas informando, primeiramente, o nome da nova coluna e qual condição ou cálculo irá retornar os dados.

```python
df['idade_proxima_decada'] = df['idade'] + 10
df
```

### Remover colunas

Para indicarmos que estamos removendo uma coluna, devemos especificar o nome da coluna juntamente com o parâmetro `axis=1`.

```python
df = df.drop('idade_proxima_decada', axis=1)

df
```

### Limpeza de dados

Quando trabalhamos com dados, não podemos ter valores nulos, uma vez que esses valores faltantes atrapalhariam nossa análise.

Pensando nisso, podemos usar o método `.dropna()` para a remoção desses dados faltantes.

Obtemos como resultado apenas os registros onde todos os campos estão preenchidos.

```python
df.dropna()
```

### Limpeza focada em apenas uma coluna.

Para uma limpeza focada em apenas uma coluna, usamos o parâmetro `subset`.

```python
df.dropna(subset= 'idade')
```

### Preenchimento de campos nulos.

Quando precisamos preencher campos nulos, podemos usar o método `.fillna()`.

```python
df = df.fillna(0)

df
```

### Convertendo para tipos inteiros

Para converter um valor para inteiro, usamos o método `.astype(int)`.

```python
df['idade'] = df['idade'].astype(int)
df
```

### Contagem de ocorrências.

Para a contagem de ocorrências, usamos o método `.value_counts()`.

```python
df['estado'].value_counts()
```
